```python
# 4. Crea un dataframe con TODOS os episodios da primeira páxina de resultados
```

```python
import pandas as pd

endpoint = 'https://rickandmortyapi.com/api/episode'
response = requests.get(endpoint, params={'page':1})
episodios = response.json()['results']
df= pd.DataFrame(episodios)
```

```python
df
#  primeira páxina=20 episodios
```

```python
# 5. Crea un dataframe con TODOS os episodios (de todas as páxinas)

# vaise crear un dataframe de cada page(en total 3) e facer un concat para unilos
```

```python
endpoint = 'https://rickandmortyapi.com/api/episode'
# dataframe paxina 1
response = requests.get(endpoint, params={'page':1})
episodios = response.json()['results']
dfpage1= pd.DataFrame(episodios)
dfpage1
```

```python
endpoint = 'https://rickandmortyapi.com/api/episode'
# dataframe paxina 2
response = requests.get(endpoint, params={'page':2})
response = response.json()['results']
dfpage2= pd.DataFrame(response)
dfpage2
```

```python
endpoint = 'https://rickandmortyapi.com/api/episode'
# dataframe paxina 3
response = requests.get(endpoint, params={'page':3})
response = response.json()['results']
dfpage3= pd.DataFrame(response)
dfpage3
```

```python
# a funcion concat() une dataframes
df_episodes=pd.concat([dfpage1,dfpage2,dfpage3])
df_episodes
```

```python
# 6. Elimina a columna dos personaxes (en caso de que exista) do dataframe que contén todos os episodios
```

```python
if 'characters' in df_episodes.columns: # se a columna dos personaxes existe no dataframe de episodios
    df_episNoCharacters=df_episodes.drop(['characters'], axis=1) # elimina a columna
df_episNoCharacters
```

```python
# 9. Crea un dataframe con todas as personaxes da serie cos seguintes datos: nome, especie e xénero
```

```python
endpoint = 'https://rickandmortyapi.com/api/character'
response = requests.get(endpoint)
characters = response.json()['results']
# paso 1: creamos o dataframe con todos os datos das personaxes
df_characters = pd.DataFrame(characters)
df_characters
# non sei como mostrar os 826 personaxes nun mesmo dataframe sen que só saian os 20 primeiros 
```

```python
# paso 2: obtemos os datos de nome, especie e xénero creando un dataframe de cada un
df_nomes=df_characters[['name']]
df_especies = df_characters[['species']]
df_xenero = df_characters[['gender']]
df_nomes
```

```python
df_especies
```

```python
df_xenero
```

```python
# paso 3: creamos o df 
df_total=pd.concat([df_nomes,df_especies,df_xenero], axis=1) 
# sen axis=1 aparecen valores nulos NaN
# axis=1 para indicar que concatene as columnas(=1) e non ao indice(=0) que ven por defecto
df_total
```
