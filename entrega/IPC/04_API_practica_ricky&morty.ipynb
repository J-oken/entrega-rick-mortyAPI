```python
# 9. Crea un dataframe con todas as personaxes da serie cos seguintes datos: nome, especie e xénero
```

```python
all_characters = []

for i in range(1, 42+1):
    url = f"https://rickandmortyapi.com/api/character?page={i}"
    response_p = requests.get(url)
    data_p = response_p.json()
    df_p = pd.DataFrame(data_p["results"])
    all_characters.append(df_p)

all_characters_df = pd.concat(all_characters, ignore_index=True)
```

```python
all_characters_df[['name', 'species', 'gender']]
```
